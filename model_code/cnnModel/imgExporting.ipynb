```python
import os
import numpy as np
from PIL import Image

# Number of images to generate
num_images = 4580

# Image dimensions (height and width)
height, width = 201, 201

# Output directory for saving images
output_directory = "generated_images"
os.makedirs(output_directory, exist_ok=True)

# Loop to generate and save images
for i in range(num_images):
    # Create a blank white image with transparency (RGBA)
    image = np.ones((height, width, 4), dtype=np.uint8) * 255

    # Add random colors and patterns
    image[:, :, 0] = np.random.randint(0, 256, size=(height, width), dtype=np.uint8)  # Red channel
    image[:, :, 1] = np.random.randint(0, 256, size=(height, width), dtype=np.uint8)  # Green channel
    image[:, :, 2] = np.random.randint(0, 256, size=(height, width), dtype=np.uint8)  # Blue channel
    image[:, :, 3] = np.random.randint(0, 256, size=(height, width), dtype=np.uint8)  # Alpha (transparency) channel

    # Convert the NumPy array to a PIL Image
    pil_image = Image.fromarray(image, mode='RGBA')

    # Save the image with a unique filename
    filename = f"image_{i}.png"
    file_path = os.path.join(output_directory, filename)
    pil_image.save(file_path)

print(f"{num_images} images saved in the '{output_directory}' directory.")
```
