```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=[]
y=[]
for i in range(4577):
   s=[]
   s.append(arrlist[i])
   s.append(arrlist[i+1])
   s.append(arrlist[i+2])
   x.append(s) 
   y.append(arrlist[i+3][0:2])
```

```python
xtrain=np.array(x)
```

```python
ytrain=np.array(y)
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from keras.layers import Dense
from keras.layers import LSTM
from keras.models import Sequential
```

```python
# create the LSTM model
model = Sequential()
from keras.layers import Input
model.add(layers.Input(shape=(3,4)))
model.add(LSTM(128,activation='LeakyReLU',return_sequences=False))
model.add(Dense(2,activation='LeakyReLU'))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(xtrain,ytrain,epochs=200)
```

```python
output=model.predict(xtrain)
ytrain[:,1]
```
